```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.model_selection import cross_val_score

df = pd.read_excel('dataBaseTratada.xlsx')

df.head()

df2 = pd.read_excel('dataBaseTeste_final.xlsx')
```

```python
df.shape
```

```python
X_df = df[['freight_value', 'Distancia', 'product_weight_g', 'product_width_cm', 'product_height_cm', 'product_length_cm']]
Y_df = df['delivery_time']

X_df.head()
```

```python
Y_df.head()
```

```python
X_df_train = X_df
Y_df_train = Y_df
```

```python
X_df_test = df2[['freight_value', 'Distancia', 'product_weight_g', 'product_width_cm', 'product_height_cm', 'product_length_cm']]
```

```python
regressor_lgbm = lgb.LGBMRegressor()
```

```python
cross_val_score(regressor_lgbm, X_df_train, Y_df_train, scoring='neg_root_mean_squared_error').mean()
```

```python
regressor_lgbm_tunado = lgb.LGBMRegressor(max_depth=2)

cross_val_score(regressor_lgbm_tunado, X_df_train, Y_df_train, scoring='neg_root_mean_squared_error').mean()
```

```python
regressor_lgbm_dart = lgb.LGBMRegressor(boosting_type='dart')

cross_val_score(regressor_lgbm_dart, X_df_train, Y_df_train, scoring='neg_root_mean_squared_error').mean()
```

```python
regressor_campeao = regressor_lgbm

regressor_campeao.fit(X_df_train, Y_df_train, eval_metric='root_mean_squared_error')
```

```python
predicoes = regressor_campeao.predict(X_df_test)

predicoes[:10]
```

```python
Y_df_test[:10]
```

```python
from sklearn.metrics import mean_squared_error
import math

mse = mean_squared_error(Y_df_train, predicoes)

display(mse)

rmse = math.sqrt(mse)

rmse
```
